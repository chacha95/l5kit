```python
from draw import draw_road_ref, rasterize

import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import numpy as np

tree = ET.parse('/dataset/opendrive/borregasave.xodr')
root = tree.getroot()

plt.figure(dpi=400)

all_points = list()
all_points = np.array(all_points)
for i, sample in enumerate(root.iter('road')):
    points = draw_road_ref(sample[2])
    plt.imshow(rasterize(points, 5))
    
    all_points = np.append(all_points, points).reshape(-1, 2)
plt.imshow(rasterize(all_points, 5))
plt.show()
```
