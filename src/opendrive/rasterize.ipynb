```python
from matplotlib import pyplot as plt
from parser import OpendriveParser

import xml.etree.ElementTree as ET
import numpy as np

map = "/dataset/opendrive/test1.xodr"
# map = "/dataset/opendrive/borregasave.xodr"

parser = OpendriveParser(map)
parser()
```

```python
size = 20
plt.figure(figsize=(size, size))

data_dict = {'data_x': parser.road[:, 0], 
             'data_y': parser.road[:, 1]}
plt.plot('data_x', 'data_y', 'ro', data=data_dict)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

min_size = min(np.reshape(parser.road, (-1)))
max_size = max(np.reshape(parser.road, (-1))) 
plt.xlim([min_size, max_size])  
plt.ylim([min_size, max_size])
plt.show()
```
